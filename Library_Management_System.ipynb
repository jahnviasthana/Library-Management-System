```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        """Add a new book to the library."""
        self.books.append({"title": title, "author": author, "borrowed": False})
        print(f'Book "{title}" by {author} added successfully.')

    def display_books(self):
        """Display all available books."""
        print("\nAvailable Books:")
        found = False
        for book in self.books:
            if not book["borrowed"]:
                print(f'- {book["title"]} by {book["author"]}')
                found = True
        if not found:
            print("No books available.")

    def borrow_book(self, title):
        """Borrow a book if it's available."""
        for book in self.books:
            if book["title"].lower() == title.lower() and not book["borrowed"]:
                book["borrowed"] = True
                print(f'You have borrowed "{book["title"]}".')
                return
        print(f'Sorry, "{title}" is not available.')

    def return_book(self, title):
        """Return a borrowed book."""
        for book in self.books:
            if book["title"].lower() == title.lower() and book["borrowed"]:
                book["borrowed"] = False
                print(f'You have returned "{book["title"]}".')
                return
        print(f'"{title}" was not borrowed or does not exist.')

# Main program loop
library = Library()

while True:
    print("\nLibrary Menu:")
    print("1. Add Book")
    print("2. View Books")
    print("3. Borrow Book")
    print("4. Return Book")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        title = input("Enter book title: ")
        author = input("Enter author name: ")
        library.add_book(title, author)

    elif choice == "2":
        library.display_books()

    elif choice == "3":
        title = input("Enter book title to borrow: ")
        library.borrow_book(title)

    elif choice == "4":
        title = input("Enter book title to return: ")
        library.return_book(title)

    elif choice == "5":
        print("Exiting Library System. Goodbye!")
        break

    else:
        print("Invalid choice, please try again.")
```
