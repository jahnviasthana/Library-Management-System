```python
import pandas as pd

# ---------------- LOAD DATA ----------------
df = pd.read_csv("Books.csv")
df['is_issued'] = df['is_issued'].astype(bool)

# ---------------- LIBRARY CLASS ----------------
class Library:
    def __init__(self, dataframe):
        self.df = dataframe

    def display_books(self, n=10):
        print("\nSample Books:")
        display(self.df.head(n))

    def add_book(self, book_id, title, author, subject):
        if book_id in self.df['book_id'].values:
            print("Book ID already exists!")
            return

        new_row = pd.DataFrame([{
            'book_id': book_id,
            'title': title,
            'author': author,
            'subject': subject,
            'is_issued': False
        }])

        self.df = pd.concat([self.df, new_row], ignore_index=True)
        print("Book added successfully!")

    def issue_book(self, book_id):
        if book_id not in self.df['book_id'].values:
            print("Book not found!")
            return

        idx = self.df[self.df['book_id'] == book_id].index[0]

        if self.df.at[idx, 'is_issued']:
            print("Book already issued!")
        else:
            self.df.at[idx, 'is_issued'] = True
            print("Book issued successfully!")

    def return_book(self, book_id):
        if book_id not in self.df['book_id'].values:
            print("Book not found!")
            return

        idx = self.df[self.df['book_id'] == book_id].index[0]

        if not self.df.at[idx, 'is_issued']:
            print("Book was not issued!")
        else:
            self.df.at[idx, 'is_issued'] = False
            print("Book returned successfully!")

    # -------- IMPROVED SEARCH (SUMMARY VIEW) --------
    def search_book(self, keyword):
        result = self.df[
            self.df['title'].str.contains(keyword, case=False, na=False) |
            self.df['author'].str.contains(keyword, case=False, na=False) |
            self.df['subject'].str.contains(keyword, case=False, na=False)
        ]

        if result.empty:
            print("No matching books found.")
        else:
            summary = result.groupby(['title', 'author']).agg(
                total_copies=('book_id', 'count'),
                available_copies=('is_issued', lambda x: (~x).sum())
            ).reset_index()

            print("\nBook Availability Summary:")
            display(summary)

    # -------- IMPROVED SUBJECT FILTER --------
    def filter_by_subject(self, subject):
        result = self.df[self.df['subject'].str.lower() == subject.lower()]

        if result.empty:
            print("No books found in this subject.")
        else:
            summary = result.groupby(['title', 'author']).agg(
                total_copies=('book_id', 'count'),
                available_copies=('is_issued', lambda x: (~x).sum())
            ).reset_index()

            print(f"\nBooks in {subject}:")
            display(summary)

    def save_data(self):
        self.df.to_csv("updated_books.csv", index=False)
        print("Data saved to updated_books.csv")

# ---------------- INITIALIZE ----------------
library = Library(df)

# ---------------- MENU ----------------
while True:
    print("\n====== Library Management System ======")
    print("1. Display Books")
    print("2. Add Book")
    print("3. Issue Book")
    print("4. Return Book")
    print("5. Search Book")
    print("6. Filter by Subject")
    print("7. Save & Exit")

    choice = input("Enter your choice: ")

    try:
        if choice == '1':
            library.display_books()

        elif choice == '2':
            book_id = int(input("Enter Book ID: "))
            title = input("Enter Title: ")
            author = input("Enter Author: ")
            subject = input("Enter Subject: ")
            library.add_book(book_id, title, author, subject)

        elif choice == '3':
            book_id = int(input("Enter Book ID: "))
            library.issue_book(book_id)

        elif choice == '4':
            book_id = int(input("Enter Book ID: "))
            library.return_book(book_id)

        elif choice == '5':
            keyword = input("Enter keyword (title/author/subject): ")
            library.search_book(keyword)

        elif choice == '6':
            subject = input("Enter subject: ")
            library.filter_by_subject(subject)

        elif choice == '7':
            library.save_data()
            print("Exiting system...")
            break

        else:
            print("Invalid choice!")

    except Exception as e:
        print("Error:", e)
```
